```python
num = 0 
while (num<5 ):
    np.random.seed(1)
    print(np.random.randint(10))
    num +=1
```

```python
num = 0 
np.random.seed(1)
while (num<5 ):
    
    print(np.random.randint(10))
    num +=1

#random seed 在外面所有只有第一過有被規範
```
